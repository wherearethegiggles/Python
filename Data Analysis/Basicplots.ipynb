```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
# Line Chart
plt.figure(figsize=(8, 6))
plt.plot(iris.sepal_length, label='Sepal Length', color='blue', linestyle='-', marker='o')
plt.plot(iris.sepal_width, label='Sepal Width', color='green', linestyle='--', marker='s')
plt.title('Line Chart: Sepal Length vs Sepal Width')
plt.xlabel('Index')
plt.ylabel('Length / Width')
plt.legend()
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(iris.sepal_length, iris.petal_length, color='red', marker='o', label='Sepal vs Petal Length')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(iris.sepal_width, bins=15, color='purple', alpha=0.7)
plt.title('Histogram: Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart
species_counts = iris.species.value_counts()
plt.figure(figsize=(8, 6))
plt.bar(species_counts.index, species_counts.values, color='orange')
plt.title('Bar Chart: Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
